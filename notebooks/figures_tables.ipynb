## (1) Tilting relative entropy
----------------------------------------------

A particular parametric alternative $(\alpha, \kappa)$ can be represented by adding the drift distortion $S_t=\frac{\alpha-\widehat\alpha}{\sigma} + \frac{\widehat\kappa-\kappa}{\sigma} (X_t-\bar x)$ to the Brownian shock. The corresponding discounted tilted relative entropy is 
\begin{align}
\varrho\left(M^S \mid x, \delta, \xi\right) : = \frac{\delta}{2}\int_{0}^{\infty}\exp(-\delta t)E^S\left[(\vert S_t\vert^2 -\xi(X_t))  \mid X_0=x\right]
\end{align}
which is a sum of terms quadratic in $X_t$, where $X$ follow linear dynamics. As such, for each $(\alpha, \kappa)$, we can express it as 
\begin{align}
\varrho(x; \delta, \xi_0, \xi_1, \xi_2) := \frac{\rho_0}{2} + \rho_1\left(\frac{\alpha}{\kappa}-\bar x\right) + \frac{\rho_2}{2}\left(\frac{ |\sigma|^2}{2\kappa} + \left(\frac{\alpha}{\kappa}-\bar{x}\right)^2\right) + \delta\left(\frac{\rho_2\left(\frac{\alpha}{\kappa} - \bar x\right) + \rho_1}{(\delta +  \kappa)}\right)\left[x-\frac{\alpha}{\kappa}\right] + \frac{\delta\rho_2}{2(\delta + 2\kappa)}\left[   \left[x-\frac{\alpha}{\kappa}\right]^2-\frac{|\sigma|^2}{2\kappa} \right]
\end{align}

where $(\widehat\alpha, \widehat\kappa)$ are the baseline parameters and 

$$\rho_2 = \left(\frac{\kappa-\hat{\kappa}}{\sigma^2}\right)^2 - \xi_2 \quad\quad\rho_1 = \left(\frac{\alpha-\hat{\alpha}}{\sigma}\right)\left(\frac{\kappa-\hat{\kappa}}{\sigma}\right) - \xi_1\quad\quad\rho_0 = \left(\frac{\alpha-\hat{\alpha}}{\sigma}\right)^2 - \xi_0$$


$\hat \alpha = 0$, $\hat \kappa = .169$, $\sigma = .195$, $\delta = .002$, $x_0 = 0$

### Instant-by-instant constraint

Suppose the constraint 

$$\left|\eta(z)\right|^2\leq\xi(z) \quad\quad\quad \forall z$$

with the definition

$$\xi(x) = \xi_0 + 2\xi_1 x + \xi_2 x^2 \ \dot = \ \left(\frac{\widetilde\alpha - \widehat\alpha}{\sigma}\right)^2 + 2\left(\frac{\widetilde\alpha - \widehat\alpha}{\sigma}\right)\left(\frac{\widehat\kappa - \widetilde\kappa}{\sigma}\right) x + \left(\frac{\widehat\kappa - \widetilde\kappa}{\sigma}\right)^2x^2.$$


>This constraint is equivalent with the requirement that a generic quadratic function 

>$$ax^2 + 2bx + c \leq 0 \quad\quad \forall x \ (!)$$

>In order for this to be satisfied, 
> * $c\leq 0$ : to take care of $x=0$
> * $a\leq 0$ : to take care of $|x|\to \infty$
> * $b^2 - ac\leq 0$ : cannot have two roots 

In our case, the parameters of the quadratic function are given by

$$a = \left(\frac{\widehat\kappa - \kappa}{\sigma}\right)^2 - \xi_2$$

$$b = \left(\frac{\alpha - \widehat\alpha}{\sigma}\right)\left(\frac{\widehat\kappa - \kappa}{\sigma}\right) - \xi_1$$

$$c = \left(\frac{\alpha - \widehat\alpha}{\sigma}\right)^2 - \xi_0$$

implying 

$$\kappa_{1, 2} = \hat\kappa \pm \sigma \sqrt{\xi_2} \quad\quad \text{and}\quad\quad \alpha_{1, 2} = \hat\alpha \pm \sigma \sqrt{\xi_0}$$

so 

 * if $\xi_1>0$, then $(\kappa_{1}, \alpha_{2})$ and $(\kappa_{2}, \alpha_{1})$
 * if $\xi_1<0$, then $(\kappa_{1}, \alpha_{1})$ and $(\kappa_{2}, \alpha_{2})$

solve the constraint with equality.

W.l.o.g., suppose $\xi_1<0$ and consider the convex combinations of the associated points with $(\hat\kappa, \hat\alpha)$

\begin{align}
\kappa = \lambda_{\kappa}\left(\hat\kappa - \sigma\sqrt{\xi_2}\right) + (1-\lambda_{\kappa})\hat\kappa = \hat\kappa - \lambda_{\kappa}\sigma\sqrt{\xi_2} \quad\quad \lambda_{\kappa}\in[-1,1] \\
\alpha = \lambda_{\alpha}\left(\hat\alpha - \sigma\sqrt{\xi_0}\right) + (1-\lambda_{\alpha})\hat\alpha = \hat\alpha - \lambda_{\alpha}\sigma\sqrt{\xi_0} \quad \quad \lambda_{\alpha}\in[-1,1] 
\end{align}

and so we have $a=(\lambda_{\kappa}^2-1)\xi_2<0$ and $c=(\lambda_{\alpha}^2-1)\xi_0<0$ and 

\begin{align}
b^2-ac &= \left(\lambda_{\kappa}\lambda_{\alpha}\sqrt{\xi_0\xi_2} - \xi_1\right)^2 - (\lambda_{\kappa}^2-1)(\lambda_{\alpha}^2-1)\xi_2\xi_0= \\
&= \left[\lambda_{\kappa}^2 + \lambda_{\alpha}^2 -1\right]\xi_0\xi_2 + \xi_1^2 - 2\lambda_{\kappa}\lambda_{\alpha}\sqrt{\xi_0\xi_2}\xi_1
\end{align}

> **Special case:** when $\sqrt{\xi_2\xi_0}=\xi_1$ (like in the case of our constructed $\xi$ functions), this is

>\begin{align}
b^2-ac &= \xi_1^2\left(\lambda_{\kappa} - \lambda_{\alpha}\right)^2 
\end{align}

> i.e. $b^2-ac\leq 0 $ iff $\lambda_{\kappa}=\lambda_{\alpha}$.

Alternatively, suppose the constraint 

$$\left|\eta(z)\right|^2\leq\xi(z)$$

$z$-by-$z$. For each $z$, this draws a (possibly tilted) ellipse in the $(\alpha, \kappa)$ space. Writing the constraint as an equality (and plugging in for $(\eta_0, \eta_1)$) shows that $z$ affects the parameters of the ellipse:

$$\frac{1}{\sigma^2\xi(z)}\left(\alpha-\widehat\alpha\right)^2 + \frac{2z}{\sigma^2\xi(z)}(\alpha-\widehat\alpha)(\widehat\kappa-\kappa) + \frac{z^2}{\sigma^2\xi(z)}(\widehat\kappa-\kappa)^2 = 1\quad\quad\text{if}\quad \xi(z)\neq 0$$


```python
def ellipse_ak(alpha, kappa, z, xi_0, xi_1, xi_2, alpha_hat=alpha_hat, kappa_hat=kappa_hat, sigma=sigma):
    xiz = sigma**2 * (xi_0 + 2*xi_1*z + xi_2*z**2)
    a = 1/xiz
    b = 2*z/xiz
    c = z**2/xiz
    
    lhs = a*(alpha-alpha_hat)**2 + b*(alpha-alpha_hat)*(kappa_hat-kappa) + c*(kappa_hat-kappa)**2 - 1
    
    return lhs
```

## Worst-case models

We revisit the example in section 3.2. Suppose that the decision maker's baseline model for the univariate state $X$ is affine and is hit by a single Brownian shock $W$

\begin{align}
dX_t = \left(\hat{\alpha} - \hat{\kappa} x\right)dt + \sigma dW_t
\end{align}

Let the period utility be $\upsilon(c)=\delta \log C$ and suppose that a decision maker contemplates a particular consumption policy $\bar d$ (indexed by $(\bar\phi, \bar \beta)$) that would lead to the consumption dynamics

\begin{align*}
d\log C_t = \left(\bar\phi + \bar \beta x\right)dt
\end{align*}

This could be viewed as a special case of the single capital model in section 5 (when $\sigma_k=0$). There, the agent can affect $\bar\phi$ but not $\bar\beta$.

* The decision maker takes $(\bar\phi, \bar \beta)$ and investigates the fragility of this policy to misspecifications in the state dynamics by computing the worst-case model. 

* She entertains tilting function $\xi$ and robustness parameter $\mathsf q$. We use quadratic $\xi$ to direct the agent's concern to persistence. 

* We can use formulas from Appendix to find that $s^*$ and $u^*$ are linear in $x$, so the worst-case has the same form as the baseline. This means that we can utilize again our projection exercise. (although the set is much bigger, we know that the worst-case will belong to the parametric class). 

* We use the projection exercise to see how different values of $(\xi, \mathsf q)$ can influence the worst-case associated with the proposed $d$.

---------------------------------------------------------------------
### Iso-utility curves in the parametric class
---------------------------------------------------------------------

The value function of the household for a particular policy $(\bar\phi, \bar\beta)$ under model $(\alpha, \kappa)$ is
\begin{align}
V(x, c) &= E^S\left[\int_{0}^{\infty}\delta\exp(-\delta t)\left(\log C_t -\log C_0\right) \mid X_0=x, \log C_0=c\right] + c \\
&= \lambda \cdot x + c
\end{align}
where
\begin{align}
\lambda=  \begin{bmatrix} \delta & -\alpha \\ 0 & \delta+\kappa \end{bmatrix}^{-1}(.01)(\bar\phi + \bar\beta x) = \frac{(.01)}{\delta(\delta+\kappa)}\begin{bmatrix} \delta + \kappa & \alpha \\ 0 & \delta \end{bmatrix}\begin{bmatrix}\bar\phi \\ \bar\beta \end{bmatrix} = (.01)\begin{bmatrix}\left(\frac{\bar\phi}{\delta} + \frac{\alpha\bar\beta}{\delta(\delta+\kappa)}\right)\\ \frac{\bar\beta}{\delta + \kappa} \end{bmatrix}
\end{align}

## Contrast sets with tenuous beliefs

---------------------------------------------------------------------------

For the "average" constraint, we need to project the set $\mathcal M^*$ onto the parametric class defined by the change of measures $S=\eta_0 + \eta_1 z$. For such $S$, the value of $\varrho$ expressed in terms of $(\eta_0, \eta_1)$ is

\begin{align}
\lim_{\delta\downarrow 0}\varrho(x) \equiv \frac{\eta_0^2-\xi_0}{2} + \frac{(\eta_0\eta_1-\xi_1)(\hat{\alpha}+\sigma\eta_0)}{(\hat{\kappa}-\sigma\eta_1)}+ \frac{(\eta_1^2 - \xi_2)}{2}\left(\frac{ |\sigma|^2}{2(\hat{\kappa}-\sigma\eta_1)} + \left(\frac{\hat{\alpha}+\sigma\eta_0}{(\hat{\kappa}-\sigma\eta_1)}\right)^2\right) \leq 0
\end{align}

with $\eta_1\leq \frac{\hat\kappa}{\sigma}$. We need the $(\eta_0, \eta_1)$ pairs on the boundary, i.e. when $\varrho=0$. Using this constraint, we can express $\eta_0$ as a function of $\eta_1$ (there are two values for each $\eta_1$). For a given $z$, the maximum and minimum $S$ can be found by 

$$\max_{\eta_1} S=\eta_0(\eta_1) + \eta_1 z \quad \text{and}\quad \min_{\eta_1} S=\eta_0(\eta_1) + \eta_1 z$$

```python
delta = .01

#----------------------------------------------
# Baseline parameters (shut down first shock)
#----------------------------------------------
observable = [.373, 1., [.0, 0.]]
state = [0., .017, [.0, .027]]

sets = SetsModel(state, observable, delta)
tilting = sets.generate_xi(.005)
sets.worst_case(tilting, theta=np.inf)
```

```python
alpha_hat = sets.alpha_z
kappa_hat = sets.kappa
sigma = sets.sigma_z[1, 0]
zbar = sets.zbar

#xi0 = .09**2
#xi2 = kappa_hat*xi0/sigma**2

xi2 = sets.generate_xi(.005)[2]
xi0 = xi2*sigma**2/kappa_hat/2

s1_sets = - np.sqrt(xi2*(z_grid-zbar)**2)
s2_sets = + np.sqrt(xi2*(z_grid-zbar)**2)

s1_tenuous = -(2*xi0/sigma)*(z_grid-zbar) - np.sqrt(((2*xi0/sigma)**2 + 4*xi2)*(z_grid-zbar)**2)
s2_tenuous = -(2*xi0/sigma)*(z_grid-zbar) + np.sqrt(((2*xi0/sigma)**2 + 4*xi2)*(z_grid-zbar)**2)
```

```python
def tilted_entropy_in_eta(eta0, eta1, xi_0, xi_1, xi_2, alpha_hat=alpha_hat, kappa_hat=kappa_hat, sigma=sigma):
    
    rho_0 = eta0**2 - xi_0
    rho_1 = eta0*eta1 - xi_1
    rho_2 = eta1**2 - xi_2
    
    alpha = alpha_hat + sigma*eta0
    kappa = kappa_hat - sigma*eta1
    
    mean = alpha/kappa
    var = sigma**2/(2*kappa)

    return rho_0/2 + rho_1*mean + (.5)*rho_2*(var + mean**2)


def S_z(eta1, z, bounds, xi_0, xi_1, xi_2, alpha_hat=alpha_hat, kappa_hat=kappa_hat, sigma=sigma):
    eta0 = opt.brentq(lambda x: tilted_entropy_in_eta(x, eta1, xi_0, xi_1, xi_2, alpha_hat, kappa_hat, sigma), 
                      bounds[0], bounds[1])
    
    return eta0 + eta1*z

#==============================================================

S_min, S_max = [], []

for z in z_grid:
    S_max.append(-opt.minimize_scalar(lambda x:  -S_z(x, z, (0, 0.15), 0, 0, xi2, alpha_hat, kappa_hat, sigma), 
                                     bounds=(-.45, 0.45), method='bounded').fun)
    S_min.append(opt.minimize_scalar(lambda x: S_z(x, z, (-0.15, 0), 0, 0, xi2, alpha_hat, kappa_hat, sigma), 
                                     bounds=(-.45, 0.45), method='bounded').fun)
```
